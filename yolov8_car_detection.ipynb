```python
import os
import shutil
from tqdm.notebook import tqdm
```

```python
train_path_img="./yolo_data/images/train/"
train_path_label="./yolo_data/labels/train/"
val_path_img="./yolo_data/images/val/"
val_path_label="./yolo_data/labels/val/"
test_path="./yolo_data/test/"
```

```python
import random
random.seed(42)
```

```python
def train_test_split(path,split=0.2):
  files=list(set([name[:-4] for name in os.listdir(path)]))
  print(f"Folder contain {len(files)} images")
  random.shuffle(files)
  test_size=int(len(files)*split)
  train_size=len(files)-test_size

  os.makedirs(train_path_img,exist_ok=True)
  os.makedirs(train_path_label,exist_ok=True)
  os.makedirs(val_path_img,exist_ok=True)
  os.makedirs(val_path_label,exist_ok=True)

  for file in tqdm(files[:train_size]):
    shutil.copy(path+file+'.png',f"{train_path_img}/"+file+'.png')
    shutil.copy(path+file+'.txt',f"{train_path_label}/"+file+'.txt')

  for file in tqdm(files[train_size:]):
    shutil.copy(path+file+'.png',f"{val_path_img}/"+file+'.png')
    shutil.copy(path+file+'.txt',f"{val_path_label}/"+file+'.txt')






```

```python
train_test_split('/content/drive/MyDrive/yolov8_car/data/')
```
